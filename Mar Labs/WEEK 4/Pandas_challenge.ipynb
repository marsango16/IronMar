```python
import pandas as pd
import numpy as np

students = pd.read_csv("student_df.csv")
students
```

```python
# Group the students by country and apply an aggregation function to each group
grouped_students = students.groupby('country').agg({
    'student_id': 'count',
    'age': 'mean',
    'has_rabbit': 'sum'
})

# Rename the columns
grouped_students = grouped_students.rename(columns={
    'student_id': 'num_students',
    'age': 'avg_age',
    'has_rabbit': 'num_students_with_rabbit'
})

# Print the resulting dataframe
print(grouped_students)
```
