```python
def bucket_states(state):
    if state == "California":
        return "West Region"
    elif state == "Oregon":
        return "North West"
    elif state == "Washington":
        return "East"
    elif state in ["Arizona", "Nevada"]:
        return "Central"
    else:
        return "Unknown"

customer_df["state"] = customer_df["state"].apply(bucket_states)

print(customer_df["state"].unique())

customer_df
```
